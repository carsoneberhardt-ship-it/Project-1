```python
import pandas as pd
```

```python
### pm2.5 and pm10 by sensor

air_data = pd.read_csv('AirQuality_Daily_StudentVersion (1).csv')

air_data = pd.DataFrame(air_data)

air_data.head(5)

stat_summary = air_data.groupby(['sensor.name']).agg(max = ('pm2.5_atm', 'max'),
                                                    min = ('pm2.5_atm', 'min'),
                                                    mean = ('pm2.5_atm', 'mean'),
                                                    median = ('pm2.5_atm', 'median'))
stat_summary

stat_summary  = air_data.groupby(['sensor.name']).agg({
    'pm2.5_atm': ['max', 'min', 'mean', 'median'],
    'pm10.0_atm': ['max', 'min', 'mean', 'median']
})

stat_summary

pivot_stats = pd.pivot_table(
    air_data,
    index = "sensor.name",
    values = ["pm2.5_atm", "pm10.0_atm"],
    aggfunc = ["max", "min", "mean", "median"]
)

pivot_stats





```

```python
## pm2.5 top5 
air_data_grouped = air_data.groupby(['sensor.name'])

air_data_pm25 = air_data_grouped.agg(max = ('pm2.5_atm', 'max'),
                                                    mean = ('pm2.5_atm', 'mean'),
                                                    median = ('pm2.5_atm', 'median'))

air_data_pm25
```

```python

top5_mean_pm25 = (
    air_data_pm25.sort_values(by = 'mean', ascending = False)
)

top5_mean_pm25.head(5)
```

```python

air_data_grouped = air_data.groupby(['sensor.name'])

air_data_pm10 = air_data_grouped.agg(max = ('pm10.0_atm', 'max'),
                                                    mean = ('pm10.0_atm', 'mean'),
                                                    median = ('pm10.0_atm', 'median'))

air_data_pm10
```

```python
top5_mean_pm10 = (
    air_data_pm10.sort_values(by = 'mean', ascending = False)
)

top5_mean_pm10.head(5)
```

```python

unhealthy_voc = air_data[air_data['voc']>=25]

unhealthy_events = unhealthy_voc[
['date', 'sensor.name', 'voc']
].sort_values('date')

unhealthy_events
```

```python

def assign_temperature(temperature):
    if temperature < 32:
        return 'Below Freezing'
    elif 32 <= temperature <= 50:
        return 'Cool'
    elif 51 <= temperature <= 70:
        return 'Warm'
    else:
        return 'Hot'

air_data['temperature_level'] = air_data['temperature'].apply(assign_temperature)
                                                             
air_data.head(5)
```

```python
air_data.groupby('temperature_level')['voc'].mean()
```

```python

def assign_Humidity(humidity):
    if humidity < 50:
        return 'Low Humidity'
    elif 50 <= humidity <= 80:
        return 'High Humidity'
    else:
        return 'Very High Humidity'
        
air_data['Humidity_level'] = air_data['humidity'].apply(assign_Humidity)
                                                             
air_data.head(5)
```

```python
air_data.groupby('Humidity_level')['voc'].mean()
```

```python
def assign_AQI5(pm25atm):
    if 101 <= pm25atm <= 150:
        return 'Unsafe for Sensitive Population'

air_data['Safety2.5'] = air_data['pm2.5_atm'].apply(assign_AQI5)

air_data.head(5)
```

```python
air_data[air_data['Safety2.5'] == 'Unsafe for Sensitive Population']
```

```python
def assign_AQI10(pm10atm):
    if 101 <= pm10atm <= 150:
        return 'Unsafe for Sensitive Population'

air_data['Safety10.0'] = air_data['pm10.0_atm'].apply(assign_AQI10)

air_data.head(5)
```

```python
air_data[air_data['Safety10.0'] == 'Unsafe for Sensitive Population']
```

```python
def assign_altitude(sensor_altitude):
    if sensor_altitude < 1000:
        return 'Low'
    elif 1000 <= sensor_altitude <= 2000:
        return 'Medium'
    else:
        return 'High'
        
air_data['altitude_level'] = air_data['sensor.altitude'].apply(assign_altitude)
                                                             
air_data
```

```python
air_data.groupby('altitude_level')['voc'].mean()
```
